```python
from numpy.core.fromnumeric import mean
first_person_correct_num = len(data.loc[data[' T/F'] == True])
first_person_total_num = len(data)

first_person_correct_rate = first_person_correct_num / first_person_total_num

first_person_mean_time = mean(data[' Time(ms)'])
first_person_mean_time

id = int(len(board_df))

board_df.loc[id] = [id, first_person_correct_rate, first_person_mean_time]
sort_num_df = board_df.sort_values(by=['correct rate','mean time'], ascending = [False, True])
sort_df = sort_num_df.reset_index(drop = True)
```
