#두 사람 승자 판별
- 파일 경로를 두 사람의 실험결과가 존재하는 파일 경로로 바꿔주어야합니다!

```python
import pandas as pd
```

```python
data1 = pd.read_csv('/content/drive/MyDrive/3-2 university/ux_psychology/Sternberg/log/고은수_여_21_아침/11-08-13-49-46_num4.txt')
data1
```

```python
from numpy.core.fromnumeric import mean
first_person_correct_num = len(data1.loc[data1[' T/F'] == True])
first_person_total_num = len(data1)

first_person_correct_rate = first_person_correct_num / first_person_total_num

first_person_mean_time = mean(data1[' Time(ms)'])
first_person_mean_time
```

```python
data2 = pd.read_csv('/content/drive/MyDrive/3-2 university/ux_psychology/Sternberg/log/권혜원_여_22_점심/11-05-15-58-55_num4.txt')
data2
```

```python
from numpy.core.fromnumeric import mean
second_person_correct_num = len(data2.loc[data1[' T/F'] == True])
second_person_total_num = len(data2)

second_person_correct_rate = second_person_correct_num / second_person_total_num

second_person_mean_time = mean(data2[' Time(ms)'])
second_person_mean_time
```

```python
print('-----------The Winner ...')

first_person_score = (first_person_correct_rate * 0.5) + (first_person_mean_time * 0.5)
second_person_score = (second_person_correct_rate * 0.5) + (second_person_mean_time * 0.5)

if first_person_score > second_person_score:
  print("The Winner is First Player!!!")
else:
  print('The Winner is Second Player!!!')
```
