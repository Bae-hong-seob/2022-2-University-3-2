# 폴더 내 모든 파일 load

- 파일경로 매번 바꿔줘야함
- 저장파일명 매번 바꿔줘야함

```python
import os
import pandas as pd
from numpy.core.fromnumeric import mean

path = "/content/drive/MyDrive/3-2 university/ux_psychology/SRT+CRT/num4"
file_lst = os.listdir(path)
# 현재 디렉토리내에 모든 파일 출력
print(file_lst) # ['test1.text','test2.text','test3.text','.ipynb_checkpoints','test4.text','test.ipynb','test5.text']
```

```python
id = 0
id_list = []
correct_rate_list = []
mean_time_list = []
```

```python
# 현재 디렉토리내에 각각의 파일을 출력
for file in file_lst:
    filepath = path + '/' + file
    print('visit'+filepath+'...')

    data = pd.read_csv(filepath)
    correct_num = len(data.loc[data[' TF'] == True])
    total_num = len(data)

    correct_rate = correct_num / total_num
    mean_time = mean(data[' Time(ms)'])

    id_list.append(id)
    correct_rate_list.append(correct_rate)
    mean_time_list.append(mean_time)

    print('-----id = ',id)
    print('-----correct_rate = ',correct_rate)
    print('-----mean_time = ', mean_time)
    print('\n')

    id+=1
```

```python
num_df = pd.DataFrame({'id' : id_list, 'correct rate' : correct_rate_list, 'mean time' : mean_time_list})
num_df
```

```python
sort_num_df = num_df.sort_values(by=['correct rate','mean time'], ascending = [False, True])
sort_num_df
```

#csv 파일명 바꿔줘야함

```python
sort_num_df.to_csv("/content/drive/MyDrive/3-2 university/ux_psychology/SRT+CRT/rankboard_num4_df.csv", index = False)
```
